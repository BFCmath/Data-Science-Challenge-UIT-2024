```python
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score
```

```python
def read_train_test_file(name_file):
    # the first row is the number of train_idx and val_idx
    # the second row is the train_idx
    # the third row is the val_idx
    
    with open(name_file, 'r') as f:
        lines = f.readlines()
        num_train, num_val = int(lines[0].split()[0]), int(lines[0].split()[1])
        train_idx = list(map(int, lines[1].split()))
        val_idx = list(map(int, lines[2].split()))
    return num_train, num_val, train_idx, val_idx

num_train, num_val, train_idx, val_idx = read_train_test_file('data\\train_val_group_split.txt')
```

```python
from sklearn.model_selection import train_test_split
def stratified_split(labels, test_size=0.1, random_state=42):
    indices = list(range(len(labels)))

    # Stratified train/val split
    train_indices, val_indices = train_test_split(
        indices, test_size=test_size, stratify=labels, random_state=random_state
    )

    return train_indices, val_indices
```

```python
import json
def read_val_submission(name_file):
    # json file
    with open(name_file, 'r') as f:
        data = json.load(f)
    return list(data["results"].values())
```

```python
import json
def read_labels(name_file):
    # json file 
    with open(name_file,'r') as f:
        data = json.load(f)
    return [data[str(id)]["label"] for id in val_idx],  [data[str(id)]["label"] for id in data.keys()]
```

```python
labels,full_label = read_labels("data\json\\vimmsd-train.json")
```

```python
def my_result(file_name,val_test_mode="group"):
    if val_test_mode == "group":
        predictions = read_val_submission(file_name)
        return f1_score(labels,predictions,average="macro")
    if val_test_mode == "stratified":
        predictions = read_val_submission(file_name)
        return f1_score([full_label[i] for i in stratified_split(full_label)[1]],predictions,average="macro")
```

```python
def count_sarcasm(file_name):
    with open(file_name, 'r') as f:
        data = json.load(f)
    count_map = {
        "not-sarcasm": 0,
        "multi-sarcasm": 0,
        "text-sarcasm": 0,
        "image-sarcasm": 0
    }
    for label in data["results"].values():
        count_map[label] += 1
    return count_map
```

```python
# add count_sarcasm to df
pre_fix_train ="train_"
pre_fix_test = "test_"
names = df["model"]
for i in range(len(names)):
    train_name = "result_test_val/"+ pre_fix_train + names[i] + ".json"
    test_name = "result_test_val/"+pre_fix_test + names[i] + ".json"
    df.loc[i,"train_not"] = count_sarcasm(train_name)["not-sarcasm"]
    df.loc[i,"train_multi"] = count_sarcasm(train_name)["multi-sarcasm"]
    df.loc[i,"train_text"] = count_sarcasm(train_name)["text-sarcasm"]
    df.loc[i,"train_image"] = count_sarcasm(train_name)["image-sarcasm"]
    df.loc[i,"test_not"] = count_sarcasm(test_name)["not-sarcasm"]
    df.loc[i,"test_multi"] = count_sarcasm(test_name)["multi-sarcasm"]
    df.loc[i,"test_text"] = count_sarcasm(test_name)["text-sarcasm"]
    df.loc[i,"test_image"] = count_sarcasm(test_name)["image-sarcasm"]
```
